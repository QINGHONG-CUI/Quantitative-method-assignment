```python
df=dg_final
def drop_column_using_vif_(df, thresh=5):
    
    while True:
        # adding a constatnt item to the data
        df_with_const = add_constant(df)

        vif_df = pd.Series([variance_inflation_factor(df_with_const.values, i) 
               for i in range(df_with_const.shape[1])], name= "VIF",
              index=df_with_const.columns).to_frame()

        # drop the const
        vif_df = vif_df.drop('const')
        
        # if the largest VIF is above the thresh, remove a variable with the largest VIF
        if vif_df.VIF.max() > thresh:
            # If there are multiple variables with the maximum VIF, choose the first one
            index_to_drop = vif_df.index[vif_df.VIF == vif_df.VIF.max()].tolist()[0]
            print('Dropping: {}'.format(index_to_drop))
            df = df.drop(columns = index_to_drop)
        else:
            # No VIF is above threshold. Exit the loop
            break

    return df
```

```python
df_predictors_selected_VIF = drop_column_using_vif_(dg_final.drop('CHO', axis=1))
print("The columns remaining after VIF selection are:")
print(df_predictors_selected_VIF.columns)
```

```python
#analyze outliers
from statsmodels.stats.outliers_influence import OLSInfluence
OLSInfluence(multi_regression_model_1).summary_frame().head()
```

```python
#detect outliers
outliers = multi_regression_model_1.get_influence()
```

```python
#HAT matrix
leverage = outliers.hat_matrix_diag
```

```python
#diffits values
dffits = outliers.dffits[0]
```

```python
#student residuals
resid_stu = outliers.resid_studentized_external
```

```python
#cook distance
cook = outliers.cooks_distance[0]
```

```python
#combine them 
contat1 = pd.concat([pd.Series(leverage, name = 'leverage'),pd.Series(dffits, name = 'dffits'),
                    pd.Series(resid_stu, name = 'resid_stu'),pd.Series(cook, name = 'cook')],axis = 1)
contat1
```

```python
#combine outliers test results with original data
dataoutliers = pd.concat([dg_final,contat1],axis = 1)
dataoutliers.head()
```

```python
#calculate the ratio of outliers
outliers_ratio = sum(np.where((np.abs(dataoutliers.resid_stu)>2),1,0))/dataoutliers.shape[0]
outliers_ratio
```

```python
#ratio of outliers is 4.8%< 5%,so delete them dierctly
#drop outliers
none_outliers = dataoutliers[~(dataoutliers['resid_stu'] >= 2)] 
none_outliers
```

```python
#refine model
multi_regression_model_2 = smf.ols(formula='CHO ~ CIP + HFP + PTA + CER + HAO',data=none_outliers).fit()

# And gives a big summary of the results:
print(multi_regression_model_2.summary())
```

```python
#drop insignificant variables and rebuild model
multi_regression_model_3 = smf.ols(formula='CHO ~ CIP + HFP',data=none_outliers).fit()

# summary the result
print(multi_regression_model_3.summary())
```

```python
beta_0, beta_1, beta_2 = multi_regression_model_3.params
rsq                    = multi_regression_model_3.rsquared
pval_0, pval_1, pval_2 = multi_regression_model_3.pvalues

print("y =", round(beta_1,3), "x1 +", round(beta_2,3), "x2 +", round(beta_0,3))
print("R^2     = ", rsq)
print("p-value_1 = ", round(pval_1,3))
print("p-value_2 = ", round(pval_2,3))
```
